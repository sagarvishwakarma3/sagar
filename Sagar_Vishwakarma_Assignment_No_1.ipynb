Q- Implement a loop that continues to ask the user for a number until they enter 'exit'.

```python
while True:
 
    user_input = input("Enter a number (or type 'exit' to quit): ")

    if user_input.lower() == 'exit':  
        print("Exiting the program.")
        break

    try:
        number = float(user_input)

        if number > 0:
            print("The number is positive.")
        elif number < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")
    except ValueError:
        print("Invalid input. Please enter a valid number or type 'exit' to quit.")
```

Q- Use break to exit the loop and continue to prompt for a new number if the input is not 'exit'

```python
while True:

    user_input = input("Enter a number (or type 'exit' to quit): ")

   
    if user_input.lower() == 'exit':  
        print("Exiting the program.")
        break  # Exits the loop

    try:
        number = float(user_input)

        if number > 0:
            print("The number is positive.")
        elif number < 0:
            print("The number is negative.")
        else:
            print("The number is zero.")
    except ValueError:
        print("Invalid input. Please enter a valid number or type 'exit' to quit.")
        continue  
```

Relational and Logical Operators

Q- Create a Python script that takes two numbers as input and prints whether both numbers are even, odd, or one of each using relational and logical operators

```python
num1 = int(input("Enter the first number: "))  
num2 = int(input("Enter the second number: ")) 

if num1 % 2 == 0 and num2 % 2 == 0:  
    print("Both numbers are even")

elif num1 % 2 != 0 and num2 % 2 != 0:  
    print("Both numbers are odd")

else:
    print("One number is even, and the other is odd")
```

Q- Write a Python program that takes an integer input and prints its binary, octal, and hexadecimal equivalents using a for loop and bitwise operators

```python
def to_binary(n):
    if n == 0:
        return '0'
    binary_str = ''
    while n > 0:
        binary_str = str(n & 1) + binary_str
        n = n >> 1
    return binary_str

def to_octal(n):
    if n == 0:
        return '0'
    octal_str = ''
    while n > 0:
        octal_str = str(n & 7) + octal_str
        n = n >> 3
    return octal_str

def to_hexadecimal(n):
    if n == 0:
        return '0'
    hex_digits = '0123456789ABCDEF'
    hexadecimal_str = ''
    while n > 0:
        hexadecimal_str = hex_digits[n & 15] + hexadecimal_str
        n = n >> 4
    return hexadecimal_str

number = int(input("Enter an integer: "))

binary_equivalent = to_binary(number)
octal_equivalent = to_octal(number)
hexadecimal_equivalent = to_hexadecimal(number)

print(f"Binary equivalent: {binary_equivalent}")
print(f"Octal equivalent: {octal_equivalent}")
print(f"Hexadecimal equivalent: {hexadecimal_equivalent}")
```
